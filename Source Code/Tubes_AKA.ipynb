```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Algoritma Rekursif
def count_vowels_recursive(text, index=0):
    vowels = "aeiouAEIOU"
    if index >= len(text):
        return 0
    return (1 if text[index] in vowels else 0) + count_vowels_recursive(text, index + 1)

# Algoritma Iteratif
def count_vowels_iterative(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

# Variabel untuk menyimpan hasil
n_values = []
recursive_times = []
iterative_times = []

print("Program menghitung jumlah huruf vokal (rekursif & iteratif)")
print("Ketik 'exit' untuk berhenti memasukkan kalimat.\n")

while True:
    text = input("Masukkan kalimat: ")
    if text.lower() == "exit":
        break

    n = len(text)
    n_values.append(n)

    # Rekursif
    start_time = time.time()
    vowel_count_rec = count_vowels_recursive(text)
    recursive_times.append(time.time() - start_time)

    # Iteratif
    start_time = time.time()
    vowel_count_itr = count_vowels_iterative(text)
    iterative_times.append(time.time() - start_time)

    print(f"Jumlah huruf vokal (Recursive): {vowel_count_rec}")
    print(f"Jumlah huruf vokal (Iterative): {vowel_count_itr}\n")

# Cetak tabel eksekusi
table = PrettyTable()
table.field_names = ["Panjang Kalimat", "Recursive Time (s)", "Iterative Time (s)"]
for i in range(len(n_values)):
    table.add_row([n_values[i], recursive_times[i], iterative_times[i]])
print(table)

# Grafik perbandingan
plt.figure(figsize=(8, 6))
plt.plot(n_values, recursive_times, label='Recursive', marker='o', linestyle='-')
plt.plot(n_values, iterative_times, label='Iterative', marker='o', linestyle='-')
plt.title('Perbandingan Waktu Eksekusi: Rekursif vs Iteratif (Count Vowels)')
plt.xlabel('Panjang Kalimat (jumlah karakter)')
plt.ylabel('Execution Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
